```python
#read vostok ice core data
import pandas as pd
df = pd.read_csv('data/ice_cores/Vostok-tpt-co2.csv', header=4)
df.dtypes
df.tail()
```

```python
df.head()
```

```python
#prep plots
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
#from matplotlib import rcParams
sns.set(font_scale=1.5, font='DejaVu Sans')
#import numpy as np
```

```python
#plot historical temperature versus time
f, ax = plt.subplots(figsize=(12, 6))
x = df.BP/1.0e5
y = df.Temp
p = ax.plot(x, y, linestyle='-')
p = ax.set_title('relative temperature over time')
p = ax.set_xlabel(r'time before present     ($10^5$ years)')
p = ax.set_ylabel('relative temp $\Delta$T    (deg C)')
```

```python
#plot historical insolation versus time
f, ax = plt.subplots(figsize=(12, 6))
x = df.BP/1.0e5
y = df.Rediation
p = ax.plot(x, y, linestyle='-')
p = ax.set_title('insolation over time')
p = ax.set_xlabel(r'time before present     ($10^5$ years)')
p = ax.set_ylabel(r'July insolation at 65$^\circ$ N    (w/m$^2$)')
```

```python
#plot historical CO2 versus time
f, ax = plt.subplots(figsize=(12, 6))
x = df.BP/1.0e5
y = df.CO2
p = ax.plot(x, y, linestyle='-')
p = ax.set_title(r'CO$_2$ abundance over time')
p = ax.set_xlabel(r'time before present     ($10^5$ years)')
p = ax.set_ylabel(r'atmospheric CO$_2$    (ppm)')
```

```python
#plot temperature vs insolation
f, ax = plt.subplots(figsize=(9, 9))
x = df.Rediation
y = df.Temp
t = df.BP
p = ax.plot(x, y, marker='o', markersize=1, linestyle='-', linewidth=0.5, label='')
idx = (t > -30.0e3)
p = ax.plot(x[idx], y[idx], marker='o', markersize=1, label='recent 30K years')
idx = (t > -10.0)
p = ax.plot(x[idx], y[idx], marker='o', markersize=10, label='present')
p = ax.set_title('relative temperature vs insolation')
p = ax.set_xlabel(r'July insolation at 65$^\circ$ N    (w/m$^2$)')
p = ax.set_ylabel('relative temp $\Delta$T    (deg C)')
p = ax.legend()
```

```python
#plot temperature vs CO2
f, ax = plt.subplots(figsize=(9, 9))
x = df.CO2
y = df.Temp
t = df.BP
p = ax.plot(x, y, marker='o', markersize=1, linestyle='-', linewidth=0.5, label='')
idx = (t > -30.0e3) & (x > 0.0)
p = ax.plot(x[idx], y[idx], marker='o', markersize=1, label='recent 30K years')
idx = (t > -2500.0) & (x > 0.0)
print t[idx]
p = ax.plot(x[idx], y[idx], marker='o', markersize=10, label='2400 yrs ago')
p = ax.set_title(r'relative temperature vs CO$_2$ abundance')
p = ax.set_xlabel(r'atmospheric CO$_2$    (ppm)')
p = ax.set_ylabel('relative temp $\Delta$T    (deg C)')
p = ax.legend()
```
